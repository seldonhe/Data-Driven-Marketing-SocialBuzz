```python
dfx = reaction_types.sort_values(by='Score',ascending =False)
dfx.set_index('Type',inplace=True)
```

```python
dfx.index
```

```python
plt.figure(figsize=(8, 4.5))
sns.countplot(x='Reaction Type', data=df, order=dfx.index)
plt.title('Distribution of Reaction Types')
plt.xticks(rotation=45)
plt.show()
```

```python
t_palette=['#c8c5ef','#b76987','#857cab']

plt.figure(figsize=(5, 3))
sns.countplot(x='Sentiment', data=df, order=df['Sentiment'].value_counts().index, palette=t_palette)
plt.title('Distribution of Sentiments')
plt.xticks(rotation=0)
plt.show()
```

```python
f_palette = ['#c8c5ef', '#b76987', '#857cab', '#f5c1d2']

# Aggregate the counts for each content type
content_type_counts = df['Content Type'].value_counts()

# Create a pie chart
plt.figure(figsize=(5, 3))
plt.pie(content_type_counts, labels=content_type_counts.index, colors=f_palette, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Content Types')

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')

plt.show()
```

```python
plt.figure(figsize=(10, 6))
category_counts = df['Category'].value_counts()
sns.barplot(x=category_counts.index, y=category_counts.values)
plt.title('Distribution of Content Categories')
plt.xticks(rotation=45)
plt.show()
```
